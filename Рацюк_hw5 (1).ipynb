1.	Для датасету px.data.iris() побудувати наступні графіки.
Графіки мають назву, що вирівнювання за центром.


1.Побудуйте 3D-графік розподілу даних для датасету Iris з використанням властивостей "sepal_length," "sepal_width," та "petal_length." Кожна точка має бути розфарбована відповідно до класу квітки (setosa, versicolor, virginica).

```python
import plotly.express as px

df_iris = px.data.iris()
fig1_1 = px.scatter_3d(df_iris,
                    x='sepal_length',
                    y='sepal_width',
                    z='petal_length',
                    color='species',
                    title='3D розподіл даних Iris')

fig1_1.update_layout(title_x=0.5)
fig1_1.show()
```

2.Побудуйте гістограму для розподілу значень "petal_width" для різних видів квіток в датасеті Iris.

```python
fig1_2 = px.histogram(df_iris,
                   x="petal_width",
                   color="species",
                   marginal="rug",
                   title="Розподіл petal_width для різних видів")

fig1_2.update_layout(title_x=0.5, barmode='overlay')
fig1_2.show()
```

3.Побудуйте лінійний графік, що відображає зміну середнього значення "sepal_length" в залежності від значення "sepal_width" для кожного виду квітки (setosa, versicolor, virginica).

```python
df_avg = df_iris.groupby(['sepal_width', 'species'], as_index=False)['sepal_length'].mean()

fig1_3 = px.line(df_avg,
              x="sepal_width",
              y="sepal_length",
              color="species",
              title="Зміна середнього sepal_length від sepal_width")

fig1_3.update_layout(title_x=0.5)
fig1_3.show()
```

2. Для датасету  px.data.gapminder() побудувати наступні графіки

1.Побудувати гістограму, яка відображає розподіл ВВП за країнами для року 2007. Обрати 10 країн з найбільшим ВВП та позначити їх на гістограмі з неперервною зміною кольору. Додати підписи над стовпцями та вирівняти їх горизонтально. Назва графіку має бути розташована за центром.

```python
df_gap = px.data.gapminder()
df_2007 = df_gap[df_gap['year'] == 2007].nlargest(10, 'gdpPercap')

fig2_1 = px.bar(df_2007,
             x='country',
             y='gdpPercap',
             color='gdpPercap',
             text='gdpPercap',
             title='ТОП-10 країн за ВВП (2007)')

fig2_1.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig2_1.update_layout(title_x=0.5, xaxis_title="Країна", yaxis_title="ВВП на душу населення")
fig2_1.show()
```

2. Побудувати кругову діаграму, яка відображає розподіл ВВП за країнами Європи у 2018 році. Виберіть 5 країн з найбільшим ВВП і підпишіть діаграму за центром, щоб вона була зручною для сприйняття.

```python
df_euro = df_gap[(df_gap['continent'] == 'Europe') & (df_gap['year'] == 2007)].nlargest(5, 'gdpPercap')

fig2_2 = px.pie(df_euro,
             values='gdpPercap',
             names='country',
             title='ТОП-5 країн Європи за ВВП')

fig2_2.update_layout(title_x=0.5)
fig2_2.show()
```

```python
df_dynamic = df_gap[(df_gap['country'] == 'Canada') | (df_gap['continent'] == 'Oceania')]

fig2_3 = px.bar(df_dynamic,
             x='year',
             y='pop',
             color='country',
             barmode='group',
             title='Динаміка населення Канади та країн Океанії')

fig2_3.update_layout(title_x=0.5)
fig2_3.show()
```

3. Побудувати лінійні графіки для тригонометричних функцій. Назву графіку вирівняти за центром

На проміжку [0;5π] функція y=2 cos⁡x(x-π/4)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5 * np.pi, 500)
y = 2 * np.cos(x - np.pi/4)

plt.figure(figsize=(10, 4))
plt.plot(x, y, label=r'$y = 2 \cos(x - \frac{\pi}{4})$', color='blue')
plt.title(r'Графік функції $y = 2 \cos(x - \frac{\pi}{4})$', loc='center')
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.legend()
plt.show()
```

На проміжку [0;2π] функція y=1/2  sin⁡(2x+π/3)

```python
x = np.linspace(0, 2 * np.pi, 500)
y = 0.5 * np.sin(2 * x + np.pi/3)

plt.figure(figsize=(10, 4))
plt.plot(x, y, label=r'$y = \frac{1}{2} \sin(2x + \frac{\pi}{3})$', color='red')
plt.title(r'Графік функції $y = \frac{1}{2} \sin(2x + \frac{\pi}{3})$', loc='center')
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.legend()
plt.show()
```

. На проміжку [-2π;2π] функція y=2tg(1/2 x+π/6)

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = 2 * np.tan(0.5 * x + np.pi/6)

y[np.abs(y) > 20] = np.nan

plt.figure(figsize=(10, 4))
plt.plot(x, y, label=r'$y = 2 tg(\frac{1}{2}x + \frac{\pi}{6})$', color='green')
plt.title(r'Графік функції $y = 2 tg(\frac{1}{2}x + \frac{\pi}{6})$', loc='center')
plt.ylim(-10, 10)
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.legend()
plt.show()
```
